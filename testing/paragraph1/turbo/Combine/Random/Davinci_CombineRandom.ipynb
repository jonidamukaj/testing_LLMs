```python
import pandas as pd

# Define the filenames of the CSV files
file_paths = ['f1_scores.csv', 'rouge_scores.csv', 'text_lengths_Random.csv', 'word2vec.csv']

# Initialize an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Iterate over each file
for file_path in file_paths:
    # Read the CSV file into a DataFrame
    data = pd.read_csv(file_path)
    
    # Get the name of the first column (assumed to be the author's name)
    author_column_name = data.columns[0]
    
    # Rename the data columns to include the file name
    data.rename(columns=lambda col: f"{author_column_name}_{col}" if col != author_column_name else col, inplace=True)
    
    # Merge the data into the combined DataFrame
    if combined_data.empty:
        combined_data = data
    else:
        combined_data = pd.merge(combined_data, data, on=author_column_name, how='outer')

# Save the combined data to a new CSV file
combined_data.to_csv('combined_metrics.csv', index=False)

print("Combined CSV file saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
csv_file = 'combined_metrics.csv'  # Replace with the path to your CSV file
df = pd.read_csv(csv_file)

# Extract relevant columns
authors = df['Name']
f1_scores_1 = df['F1 Score 1']
rouge_text_1 = df['ROUGE_text_1']
vec1 = df['vec1']


# Create a line plot
plt.figure(figsize=(10, 6))

# Plotting F1 Scores
plt.plot(authors, f1_scores_1, marker='o', label='F1 Score 1')


# Plotting ROUGE Scores
plt.plot(authors, rouge_text_1, marker='o', label='ROUGE Text 1')


# Plotting ExLength and Vectors

plt.plot(authors, vec1, marker='o', label='vec1')


# Customize the plot
plt.xlabel('Authors')
plt.ylabel('Metric Values')
plt.title('Comparison of Metrics for first generated text')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
